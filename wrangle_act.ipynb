```python
#importing libriaries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import requests
%matplotlib inline
```

### WeRateDogs Twitter Archive

```python
#creating the first data frame of enhanced archive given 
df_archive= pd.read_csv('twitter-archive-enhanced.csv')
```

### Request Image Prediction

```python
url= 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
with open('images.tsv','wb')as file:
    img=requests.get(url, auth =('user','pass'))
    file.write(img.content)
```

```python
#creating the second data frame 
df_img_predictions= pd.read_csv('images.tsv',sep='\t')
```

### Twitter API

```python
#using the text file to get data
list_json=[]
#turning the text file into JSON and creating a list 
with open('tweet-json.txt', 'r') as file:
    for line in file:
        tweet=json.loads(line)
        tweet_id=tweet['id']
        retweet_count= tweet['retweet_count']
        favorite_count= tweet['favorite_count']
        list_json.append({'tweet_id': tweet_id,
                        'retweet_count': retweet_count,
                        'favorites_count': favorite_count})
  
```

```python
#viewing the list to make sure the outcome is as needed
list_json
```

```python
#creating the 3rd data frame
df_json= pd.DataFrame(list_json, columns = ['tweet_id', 'retweet_count', 'favorites_count'])

```

# Clean

```python
clean_arch= df_archive.copy()
clean_img= df_img_predictions.copy()
clean_json= df_json.copy()
```

**Define**
- datatype of timestamp

**code**

```python
clean_arch['timestamp']=pd.to_datetime(clean_arch['timestamp'])
clean_arch['retweeted_status_timestamp']=pd.to_datetime(clean_arch['retweeted_status_timestamp'])
```

**test**

```python
clean_arch.info()
```

**Define**
- inconsistant representation of missing values

**code**

```python
# inconsistent representation of NaNs
clean_arch=clean_arch.replace('None',np.NaN)
```

**test**

```python
#test
clean_arch.head()
```

```python
clean_arch.info()
```

**Define**
- drop rettweets and replies

**code**

```python
#drop non-original tweets
clean_arch= clean_arch[clean_arch['in_reply_to_status_id'].isnull()]
clean_arch= clean_arch[clean_arch['retweeted_status_id'].isnull()]
clean_arch=clean_arch.dropna(subset=['expanded_urls'],axis=0)
```

**test**

```python
clean_arch.shape
```

**Define**
- drop extra columns

**code**

```python
#drop extra columns
clean_arch.drop(columns=['in_reply_to_status_id', 'in_reply_to_user_id',
         'retweeted_status_id', 'retweeted_status_user_id',
       'retweeted_status_timestamp', 'expanded_urls'],inplace=True)
```

**test**

```python
clean_arch.shape
```

**Define**
- changing datatype of id to str

**code**

```python
# changing tweet_id to str in all dataframes
clean_arch['tweet_id'] =clean_arch['tweet_id'].astype(str)
clean_img['tweet_id'] =clean_img['tweet_id'].astype(str)
clean_json['tweet_id']=clean_json['tweet_id'].astype(str)
```

**Define**
- change inaccurate dogs' names

**coode**

```python
#fixing dogs names
not_name=[ 'a', 'an','one', 'very', 'my', 
          'not','such', 'just', 'this', 'all', 'the', 'by']
for word in not_name:
    clean_arch['name'].replace(word,np.NaN,inplace=True)
```

**Define**
- values used as columns names

**code**

```python
#merging dog stages in one col
dog_stage=['doggo','floofer','pupper','puppo']
```

```python
for word in dog_stage:
    clean_arch[word]=clean_arch[word].replace(np.NaN,'')
```

```python
clean_arch['dog_stage']=clean_arch.doggo.str.cat(clean_arch.floofer).str.cat(clean_arch.pupper).str.cat(clean_arch.puppo)
```
